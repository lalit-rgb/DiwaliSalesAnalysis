##### Dealing with Null values
Actually the null values are only 12 in the Amount column which is not that important in a data of 11251 but still here I tried to fill the null value by creating the group of that product and then replacing the null value by no. of such product ordered * avg sold value of that product 

```python
na_amt = df[df['Amount'].isnull()]
na_amt
```

```python
pg = df.groupby('Product_ID')
```

```python
len(pg)
```

```python
p_P00273442 = pg.get_group('P00273442')
```

```python
p_P00273442
```

```python
amount = p_P00273442[p_P00273442['Amount'].notnull()]['Amount'].sum()
amount
```

```python
Qnt = p_P00273442[p_P00273442['Amount'].notnull()]['Orders'].sum()
Qnt
```

```python
for x , y in pg:
    globals()[f'p_{x}'] = pg.get_group(x)
```

```python
na_prod_list = list(na_amt['Product_ID'])
single_price = []
for x,y in pg:
    if x in na_prod_list:
        amount = globals()[f'p_{x}'][globals()[f'p_{x}']['Amount'].notnull()]['Amount'].sum()
        Qnt = globals()[f'p_{x}'][globals()[f'p_{x}']['Amount'].notnull()]['Orders'].sum()
        single_piece_price = amount/Qnt
        single_price.append(single_piece_price)
    else:
        continue
i=0
single_price[0] = 76474.0/19
single_price
```

```python
i=0
for x in na_prod_list:
    print(x, single_price[i])
    i=i+1
```

```python
i=0
for x in range(0,12):
    na_amt['Amount'] = na_amt['Orders'] * single_price[i]
    i=i+1
```

```python
na_amt
```

```python
df[df['Amount'].isnull()] = na_amt
```

```python
uniq_state = df['State'].unique()
spending_per_state=[]
top_state={}
for x in uniq_state:
    s = df[df['State'] == x]['Amount'].sum()
    o = df[df['State'] == x]['Orders'].sum()
    amt = s/o
    spending_per_state.append(amt)
    print('In State',x,'per person has spent',amt, 'this much of money' )
    top_state[x]=amt
spending_per_state.sort(reverse=True)
print('top 5 states as per avg spending rate per person are', spending_per_state[0:5])
topfive = sorted(top_state.items(), key=lambda x:x[1], reverse=True)
print(topfive[0:5])
```

```python
productot = {}
for x in df['Product_Category'].unique():
    prod_zone = {}
    for y in df['Zone'].unique():
        tot = df[(df['Product_Category'] == x) & (df['Zone'] == y)]['Amount'].sum()
        prod_zone[y] = tot
    order = sorted(prod_zone.items(), key=lambda x:x[1], reverse=True)
    productot[x] = order
for x in productot:
    print(x, productot[x][0])
```
